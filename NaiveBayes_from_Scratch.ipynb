```python
data=[
    ['First','male','s1','pass'],
    ['First','female','s2','pass'],
    ['third','male','s2','fail'],
    ['second','female','s3','fail'],
    ['First','male','s1','pass'],
    ['second','male','s3','fail'],
]
```

```python
def training_NB(data):
    total=len(data)
    pass_count=0
    fail_count=0
    for col in data:
        if col[3]=='pass':
            pass_count+=1
        else:
            fail_count+=1
    p_pass=pass_count/total
    p_fail=fail_count/total
    return p_pass,p_fail,pass_count,fail_count
```

```python
training_NB(data)
```

```python
def predict_using_NB(data,query_point):
    p_pass,p_fail,pass_count,fail_count=training_NB(data)
    second_pass=male_pass=s1_pass=0
    second_fail=male_fail=s1_fail=0
    for row in data:
        if row[3]=='pass':
            if row[0]==query_point[0]:
                second_pass+=1
            if row[1]==query_point[1]:
                male_pass+=1
            if row[2]==query_point[2]:
                s1_pass+=1
        else:
            if row[0]==query_point[0]:
                second_fail+=1
            if row[1]==query_point[1]:
                male_fail+=1
            if row[2]==query_point[2]:
                s1_fail+=1
    p_second_pass=second_pass/pass_count
    p_male_pass=male_pass/pass_count
    p_s1_pass=s1_pass/pass_count

    p_second_fail=second_fail/fail_count
    p_male_pass=male_fail/fail_count
    p_s1_pass=s1_fail/fail_count

    pass_prob=p_pass*p_second_pass*p_male_pass*p_s1_pass
    fail_prob=p_fail*p_second_fail*p_male_pass*p_s1_pass

    if pass_prob>fail_prob:
        return "Pass"
    else:
        return "fail"
```

```python
query_point=['second','male','s1']
predict_using_NB(data,query_point)
```

```python
import pandas as pd

data = [
    ["Sunny", "Hot", "High", "Weak", "No"],
    ["Sunny", "Hot", "High", "Strong", "No"],
    ["Overcast", "Hot", "High", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Strong", "No"],
    ["Overcast", "Cool", "Normal", "Strong", "Yes"],
    ["Sunny", "Mild", "High", "Weak", "No"],
    ["Sunny", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "Normal", "Weak", "Yes"],
    ["Sunny", "Mild", "Normal", "Strong", "Yes"],
    ["Overcast", "Mild", "High", "Strong", "Yes"],
    ["Overcast", "Hot", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Strong", "No"]
]
```

```python
def training_NB(data):
    total=len(data)
    yes_count=0
    no_count=0
    for col in data:
        if col[4]=='Yes':
            yes_count+=1
        else:
            no_count+=1
    p_yes=yes_count/total
    p_no=no_count/total
    return p_yes,p_no,yes_count,no_count
```

```python
training_NB(data)
```

```python
def predict_using_NB(data,query_point):
    p_yes,p_no,yes_count,no_count=training_NB(data)
    rain_yes=mild_yes=normal_yes=weak_yes=0
    rain_no=mild_no=normal_no=weak_no=0
    for row in data:
        if row[4]=='yes':
            if row[0]==query_point[0]:
                rain_yes+=1
            if row[1]==query_point[1]:
                mild_yes+=1
            if row[2]==query_point[2]:
                normal_yes+=1
            if row[3]==query_point[3]:
                weak_yes+=1
        else:
            if row[0]==query_point[0]:
                rain_no+=1
            if row[1]==query_point[1]:
                mild_no+=1
            if row[2]==query_point[2]:
                normal_no+=1
            if row[3]==query_point[3]:
                weak_no+=1
        p_rain_yes=rain_yes/yes_count
        p_mild_yes=mild_yes/yes_count
        p_normal_yes=normal_yes/yes_count
        p_weak_yes=weak_yes

        p_rain_no=rain_no/no_count
        p_mild_no=mild_no/no_count
        p_normal_no=normal_no/no_count
        p_weak_no=weak_no/no_count

    yes_prob=p_yes*p_rain_yes*p_mild_yes*p_normal_yes*p_weak_yes
    no_prob=p_no*p_rain_no*p_mild_no*p_normal_no*p_weak_no

    if yes_prob>no_prob:
        return "Yes"
    else:
        return "No"
query_point=['Rain','Mild','Normal','Weak']
predict_using_NB(data,query_point)
```
