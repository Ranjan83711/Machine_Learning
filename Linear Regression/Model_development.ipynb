**Model Development**

- split the data into two parts X and y
- Part 1: X_train and y_train
- Part 2: X_test and y_test

```python
X=df.drop('MedHouseVal',axis=1)
y=df['MedHouseVal']
```

- Sklearn 
    - ModelSelection
      - train_test_Split
- It expects both X and y
- test split = 0.2 (2-% of data become test data)
- shuffle=True (data will be generated randomly)
- random_state=1234

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1234)
```
