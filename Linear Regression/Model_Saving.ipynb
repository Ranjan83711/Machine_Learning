```python
path=r"C:\Users\Hp\Documents\NareshIT\Machine Learning\Linear Regression\California_updated.csv"
```

```python
import sklearn 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv(path)
```

```python
df.head()
```

```python
X=df.drop('MedHouseVal',axis=1)
y=df['MedHouseVal']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score
```

```python
MSE=mean_squared_error(y_test,y_pred)
RMSE=root_mean_squared_error(y_test,y_pred)
R2=r2_score(y_test,y_pred)
```

```python
print('MSE :',MSE)
print('RMSE :',RMSE)                                             
print('R2 :',R2)
```

## Model Saving
- we can save model in two ways :
    - using pickle: .pkl
    - using joblib: .joblib

**pickle**

```python
import pickle
```

- dump
    - we will save the model in .pkl format
  
    - it expects the model and location
 
  
    - it is similar to file open in write format
 

    - here also we will open the file with write bytes format

- load

  
    - similar way it needs only file name in read bytes format 

```python
# our model name is : lr
# where i want to save : same directory 
# on what name i want to save : california 
# on what extension i want to save : .pkl
# file name : california.pkl
# write format is : wb
```

```python
str1='hello'
with open('dummy.txt','w') as file:
    file.write(str1)
```

```python
with open('california.pkl','wb') as file:
    pickle.dump(lr,file)
```

```python
with open('california.pkl','rb') as file:
    loaded_file=pickle.load(file)
```

```python
loaded_file
```

## joblib

```python
import joblib 
```

```python
file='california.joblib'
joblib.dump(lr,file)
```

```python
loaded_model1=joblib.load(file)
```

```python
loaded_model1
```

```python
loaded_model1.predict([[10,20,30,40,50,60]])
```
