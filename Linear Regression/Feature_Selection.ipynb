- VIF on only input columns

```python
X=california_df.drop('MedHouseVal',axis=1)
```

```python
y=california_df['MedHouseVal']
```

```python
X
```

```python
y
```

```python
X.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# we need to pass the array of all input data
for i in range(len(X.columns.tolist())):
    print(variance_inflation_factor(X.values,i))
```

```python
lst=[]
for i in range(len(X.columns.tolist())):
    lst.append(variance_inflation_factor(X.values,i))
```

```python
lst
```

```python
vif_df=pd.DataFrame(lst,index=X.columns,columns=['VIF'])
```

```python
vif_df.sort_values(by='VIF',ascending=False)
```

```python
final_cols=['MedInc','HouseAge','Population','AveOccup','AveRooms','Latitude']
final_df=pd.DataFrame()
final_df=X[final_cols].copy()
final_df['MedHouseVal']=y
final_df
```

```python
X=final_df.drop('MedHouseVal',axis=1)
y=final_df['MedHouseVal']
```

```python
from sklearn.preprocessing import StandardScaler
```
