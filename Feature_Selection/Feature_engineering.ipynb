# Forward feature selection & Backward feature selection

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"E:\Downloads\diet_recommendation_dataset_1000 - diet_recommendation_dataset_1000.csv")
```

```python
df
```

```python
x=df.drop('Diet',axis=1)
```

```python
y=df['Diet']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
ss=StandardScaler()
```

```python
le=LabelEncoder()
```

```python
xtrain_scaled=ss.fit_transform(xtrain)
```

```python
xtest_scaled=ss.fit(xtest)
```

```python
ytrain_le=le.fit_transform(ytrain)
```

```python
ytest_le=le.fit_transform(ytest)
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
forwardfs = SequentialFeatureSelector(
    lr, 
    n_features_to_select=3, 
    cv=3,
    scoring='accuracy',
    direction='forward'
)
```

```python
forwardfs.fit(xtrain_scaled, ytrain_le)
```

```python
selected_features = forwardfs.get_support(indices=True)
```

```python
selected_features
```

```python
# backward
```

```python
backwardfs = SequentialFeatureSelector(
    lr, 
    n_features_to_select=3, 
    cv=3,
    scoring='accuracy',
    direction='backward'
)
```

```python
backwardfs.fit(xtrain_scaled, ytrain_le)
```

```python
selected_features = backwardfs.get_support(indices=True)
```

```python
selected_features
```

## recursive feature selection

```python
from sklearn.feature_selection import RFE
```

```python
selector = RFE(lr, n_features_to_select=5, step=1)
```

```python
selector.fit(xtrain_scaled, ytrain_le)
```

```python
selected_features = selector.get_support(indices=True)
```

```python
selected_features
```

## Recursive with CV

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold

min_features_to_select = 1  # Minimum number of features to consider
cv = StratifiedKFold(5)

rfecv = RFECV(
    estimator=lr,
    step=1,
    cv=cv,
    scoring="accuracy",
    min_features_to_select=min_features_to_select,
    n_jobs=2,
)
```

```python
rfecv.fit(xtrain_scaled, ytrain_le)
```

```python
selected_features = rfecv.get_support(indices=True)
```

```python
selected_features
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    key: value
    for key, value in rfecv.cv_results_.items()
    if key in ["n_features", "mean_test_score", "std_test_score"]
}
cv_results = pd.DataFrame(data)
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Mean test accuracy")
plt.errorbar(
    x=cv_results["n_features"],
    y=cv_results["mean_test_score"],
    yerr=cv_results["std_test_score"],
)
plt.title("Recursive Feature Elimination \nwith correlated features")
plt.show()
```
